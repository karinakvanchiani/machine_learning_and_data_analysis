```python
import numpy as np
from scipy.stats import laplace
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
k = 10
sample = laplace.rvs(k, size = 1000)
```

непрерывное распределение Лапласа и плостность его распределения

```python
plt.hist(sample, normed = True, bins = 20, alpha = 0.5)
plt.ylabel('number of samples')
plt.xlabel('x')
left = laplace.ppf(0.01, k)
right = laplace.ppf(0.99, k)
x = np.linspace(left, right, 100)
plt.plot(x, laplace.pdf(x, k), 'r-', lw = 3, alpha = 0.7, label = 'laplace ppf')
```

генерация 1000 выборок и построение графиков

```python
def laplace_f(nSamples, Ex, Vx):
    n = nSamples
    values = np.array([laplace.rvs(k, size = n) for x in range(1000)])
    meanVal = values.mean(axis = 1)
    plt.hist(meanVal, alpha = 0.5, normed = True)
    mu = Ex
    sigma = math.sqrt(Vx / n)
    print('Mx = ', mu, 'sigma = ', sigma)
    norm = sts.norm(loc = mu, scale = sigma)
    left = laplace.ppf(0.01, k)
    right = laplace.ppf(0.99, k)
    x = np.linspace(left, right, 100)
    pdf = norm.pdf(x)
    plt.plot(x, pdf, 'r-', lw = 3, alpha = 0.7, label = 'laplace ppf')
    plt.ylabel('samples')
    plt.xlabel('x')
```

вычисленные математическое ожидание, стандартное отклонение и дисперсия

```python
Ex = laplace.mean(k)
std = laplace.std(k)
Vx = std ** 2
```

построение графиков для выборок размера 5, 10 и 50

```python
laplace_f(5, Ex, Vx)
```

```python
laplace_f(10, Ex, Vx)
```

```python
laplace_f(50, Ex, Vx)
```

вывод: при большем значении размера выборки, точность аппроксимации увеличивается
