```python
import pandas as pd
import numpy as np

import statsmodels.stats.multitest as smm
from scipy import stats
```

```python
data = pd.read_csv('C:\\Users\\kvanc\\coursera\\files\\gene_high_throughput_sequencing.csv')
data.head()
```

```python
print(set(data['Diagnosis']))
```

```python
normal = data[data['Diagnosis'] == 'normal'].drop(['Patient_id', 'Diagnosis'], axis = 1)
early = data[data['Diagnosis'] == 'early neoplasia'].drop(['Patient_id', 'Diagnosis'], axis = 1)
cancer = data[data['Diagnosis'] == 'cancer'].drop(['Patient_id', 'Diagnosis'], axis = 1)
data_columns = data.columns.drop(['Patient_id', 'Diagnosis'])
print('normal shape: ', normal.shape)
print('early shape: ', early.shape)
print('cancer shape: ', cancer.shape)
```

```python
for column in data_columns:
    print('normal ' + column + ' p-value', stats.shapiro(normal[column]))
    print('early ' + column + ' p-value', stats.shapiro(early[column]))
    print('cancer ' + column + ' p-value', stats.shapiro(cancer[column]))
```

```python
def compare_gene(right, left, columns):
    return [stats.ttest_ind(right[column], left[column], equal_var = False).pvalue for column in columns]

def write_answer_1(ans):
    with open('answer1', 'w') as file:
        file.write(str(ans))
        
def write_answer_2(ans):
    with open('answer2', 'w') as file:
        file.write(str(ans))
```

```python
normal_early_diff_genes = compare_gene(normal, early, data_columns)
normal_early_diff_genes_count = len(list(filter(lambda pvalue: pvalue < 0.05, normal_early_diff_genes)))
print(normal_early_diff_genes_count)

early_cancer_diff_genes = compare_gene(early, cancer, data_columns)
early_cancer_diff_genes_count = len(list(filter(lambda pvalue: pvalue < 0.05, early_cancer_diff_genes)))
print(early_cancer_diff_genes_count)

write_answer_1(normal_early_diff_genes_count)
write_answer_2(early_cancer_diff_genes_count)
```

```python
(reject_n_e, normal_early_holm, alphac_n_e_sidak, alphac_n_e_bonf) = smm.multipletests(normal_early_diff_genes, 
                                                                                       alpha = 0.025, method = "holm")
normal_early_holm_count = len(list(filter(lambda pvalue: pvalue < 0.05, normal_early_holm)))
print(normal_early_holm_count)

(reject_n_e, early_cancer_holm, alphac_n_e_sidak, alphac_n_e_bonf) = smm.multipletests(early_cancer_diff_genes, 
                                                                                       alpha = 0.025, method = "holm")
early_cancer_holm_count = len(list(filter(lambda pvalue: pvalue < 0.05, early_cancer_holm)))
print(early_cancer_holm_count)
```

```python
def fold_change(control, treatment, column):
    control_mean = control[column].mean()
    treatment_mean = treatment[column].mean()
    return abs(treatment_mean / control_mean if treatment_mean > control_mean else -control_mean / treatment_mean)

def calculate_fold_changes(control_data, treatment_data, pvalues, columns, alpha):
    return map(lambda column, pvalue: fold_change(control_data, treatment_data, column), 
               filter(lambda column, pvalue: pvalue < alpha, zip(columns, pvalues)))

def write_answer_3(ans):
    with open('answer3', 'w') as file:
        file.write(str(ans))
        
def write_answer_4(ans):
    with open('answer4', 'w') as file:
        file.write(str(ans))
```

```python
folds_normal_early = calculate_fold_changes(normal, early, normal_early_holm, data_columns, 0.025)
folds_normal_early_meaningful_count = len(list(filter(lambda fld: fld > 1.5, folds_normal_early)))
print(folds_normal_early_meaningful_count)

folds_early_cancer = calculate_fold_changes(early, cancer, early_cancer_holm, data_columns, 0.025)
folds_early_cancer_meaningful_count = len(list(filter(lambda fld: fld > 1.5, folds_early_cancer)))
print(folds_early_cancer_meaningful_count)

write_answer_3(folds_normal_early_meaningful_count)
write_answer_4(folds_early_cancer_meaningful_count)
```
